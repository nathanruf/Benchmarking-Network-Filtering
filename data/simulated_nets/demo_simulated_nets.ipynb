```python
import os
import pickle
import networkx as nx
import matplotlib.pyplot as plt

def load_graph(filename):
	"""
	Load a graph from a pickle file.

	Args:
		filename (str): Name of the file to load the graph from.

	Returns:
		networkx.Graph: Loaded graph.
	"""
	with open(filename, 'rb') as f:
		return pickle.load(f)

def plot_graph(G, title):
	"""
	Plot a graph using NetworkX and Matplotlib.

	Args:
		G (networkx.Graph): Graph to plot.
		title (str): Title for the plot.
	"""
	plt.figure(figsize=(10, 8))
	pos = nx.spring_layout(G)
	nx.draw(G, pos, node_color='lightblue', node_size=500, with_labels=True, font_size=8, font_weight='bold')
	plt.title(title)
	plt.axis('off')
	plt.tight_layout()
	plt.show()


```

```python
# Get all pickle files in the current directory
pickle_files = [f for f in os.listdir('.') if f.endswith('.pickle')]

# Load and plot each graph
for file in pickle_files:
	G = load_graph(file)
	plot_title = f"Graph from {file}"
	plot_graph(G, plot_title)
	print(f"Plotted graph from {file}")
```
